# **deep learning**

building the deeplearning model


```python
train = data.take(train_length)
val = data.skip(train_length).take(val_length)
test = data.skip(train_length + val_length).take(test_length)
```

```python
len(test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python
# Conv Layer 1
model.add(Conv2D(16, (3, 3), strides=1, activation='relu', input_shape=(256, 256, 3)))
model.add(MaxPooling2D())

# Conv Layer 2
model.add(Conv2D(32, (3, 3), strides=1, activation='relu'))
model.add(MaxPooling2D())

# Conv Layer 3
model.add(Conv2D(16, (3, 3), strides=1, activation='relu'))
model.add(MaxPooling2D())

# Flatten and Dense Layers
model.add(Flatten())
model.add(Dense(256, activation='relu'))

# Output layer for 4 classes
model.add(Dense(4, activation='softmax'))

```

```python
model.compile(optimizer="adam",
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=["accuracy"])
```

```python
model.summary()
```

# Train

```python
logdir = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train,epochs=20,validation_data=val,callbacks=[tensorboard_callback])
```

```python
import numpy as np
import tensorflow as tf
from sklearn.metrics import precision_score, recall_score, accuracy_score

y_true_all = []
y_pred_all = []

for batch in test.as_numpy_iterator():
    X, y = batch
    yhat = model.predict(X)
    yhat = np.argmax(yhat, axis=1)


    y_true_all.extend(y)
    y_pred_all.extend(yhat)

precision = precision_score(y_true_all, y_pred_all, average='macro')
recall = recall_score(y_true_all, y_pred_all, average='macro')
accuracy = accuracy_score(y_true_all, y_pred_all)

print(f"Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}")
```

```python

IMAGE1_PATH = '/content/car_test2.jpg'
IMAGE2_PATH = '/content/test_bike2.jpg'
IMAGE3_PATH = '/content/test_cat2.jpg'
IMAGE4_PATH = '/content/test_dog2.jpg'
IMG_SIZE = (256, 256)
CLASS_NAMES = ['bike', 'car', 'cat', 'dog']


img1 = cv2.cvtColor(cv2.resize(cv2.imread(IMAGE1_PATH), IMG_SIZE), cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(cv2.resize(cv2.imread(IMAGE2_PATH), IMG_SIZE), cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(cv2.resize(cv2.imread(IMAGE3_PATH), IMG_SIZE), cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(cv2.resize(cv2.imread(IMAGE4_PATH), IMG_SIZE), cv2.COLOR_BGR2RGB)


fig, ax = plt.subplots(ncols=4, figsize=(20, 20))
ax[0].imshow(img1)
ax[0].title.set_text("Car")

ax[1].imshow(img2)
ax[1].title.set_text("Bike")

ax[2].imshow(img3)
ax[2].title.set_text("Cat")

ax[3].imshow(img4)
ax[3].title.set_text("Dog")

plt.show()
```

```python
IMAGE_PATHS = [
    '/content/car_test2.jpg',
    '/content/test_bike2.jpg',
    '/content/test_cat2.jpg',
    '/content/test_dog2.jpg'
]

for i, path in enumerate(IMAGE_PATHS):

    img = cv2.imread(path)
    img_resized = cv2.resize(img, IMG_SIZE)
    img_rgb = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)
    img_scaled = img_rgb / 255.0


    yhat = model.predict(np.expand_dims(img_scaled, axis=0))
    yhat_index = np.argmax(yhat, axis=1)[0]

    # Print prediction
    print(f"Image {i+1}: {path}")
    print(f"Predicted class index: {yhat_index}")
    print(f"Predicted class name: {CLASS_NAMES[yhat_index]}\n")
```
